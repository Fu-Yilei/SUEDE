```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
all_strains = pd.read_csv("all_strains.csv", index_col=0)
non_tb_complete_genomes_strains = pd.read_csv("non_tb_complete_genomes_strains.csv", index_col=0)
complete_genomes = pd.read_csv("complete_genomes.csv", index_col=0)
bovis_complete_genomes_strains = pd.read_csv("bovis_complete_genomes_strains.csv", index_col=0)
unique_to_tb = pd.read_csv("unique_to_tb.csv", index_col=0)
strain_16S = pd.read_csv("16S.csv", index_col=0)
```

```python
def sortbyMUMlength(form):
    form['length'] = form.index.str.len()
    form.sort_values('length', ascending=False, inplace=True)
    return(form.drop(["length"], axis=1))
```

```python
unique_to_tb_sorted = sortbyMUMlength(unique_to_tb)
all_strains_sorted = sortbyMUMlength(all_strains)
lengthlist = []
for i in unique_to_tb.index:
    lengthlist.append(len(i))
```
